# Eficiência da Irrigação da Cana-de-Açúcar

**Cultura:** Cana-de-açúcar (*Saccharum officinarum*)

**Sistema:** Gotejamento

**Objetivo:** Avaliar a eficiência do manejo hídrico durante 30 dias.

```
import matplotlib.pyplot as plt
agua=[24,25,23,24,26,25,24,23,24,25,26,24,23,24,25,24,26,25,24,23,24,25,24,23,24,25,26,24,23,24]
total=sum(agua)
media=total/len(agua)
print('Total aplicado:',total,'mm')
print('Média diária:',round(media,2),'mm')
plt.figure(figsize=(9,4))
plt.plot(range(1,31),agua,marker='o')
plt.title('Lâmina diária de irrigação')
plt.xlabel('Dia'); plt.ylabel('mm')
plt.grid()
plt.show()
plt.figure(figsize=(9,4))
plt.fill_between(range(1,31),agua)
plt.title('Consumo acumulado visual')
plt.show()
```

# Interpretação

Os dados indicam elevada regularidade na aplicação de água ao longo dos 30 dias de monitoramento, característica desejável para a cana-de-açúcar. A estabilidade da lâmina aplicada reduz riscos de déficit hídrico e perdas por percolação profunda, aumentando a eficiência do uso da água. O monitoramento contínuo possibilita ajustes precisos no manejo irrigado e contribui para maior sustentabilidade, produtividade e racionalização dos recursos hídricos.
